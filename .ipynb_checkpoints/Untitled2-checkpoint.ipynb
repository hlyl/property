```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
db_path = 'radius_database.db'

engine = create_engine(f'sqlite:///{db_path}')

query = 'SELECT * FROM property;'
df = pd.read_sql_query(query, engine)
df = df[df['sold'] != 1]
```

```python
import folium
import pandas as pd
import ipywidgets as widgets
import datetime as dt
import panel as pn

pn.extension()
from IPython.display import display

df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
df['observed'] = pd.to_datetime(df['observed'], format='%Y-%m-%d')

# Create 5 price categories
df['price_category'] = pd.qcut(df['price'], 5, labels=['very_low', 'low', 'medium', 'high', 'very_high'])

# Assuming you have already loaded the DataFrame 'df'
def update_map(min_price, max_price):
    f_df = df[(df['observed'] >= min_price) & (df['observed'] <= max_price)]

    def get_color(price_category):
        color_map = {
            'very_low': 'green',
            'low': 'lightblue',
            'medium': 'blue',
            'high': 'darkblue',
            'very_high': 'red'
        }
        return color_map[price_category]

    # Create a map centered around the average latitude and longitude of the data
    avg_latitude = f_df['latitude'].mean()
    avg_longitude = f_df['longitude'].mean()
    m = folium.Map(location=[avg_latitude, avg_longitude], zoom_start=10, width='80%', height=600)


    # Add markers for each property
    for index, row in f_df.iterrows():
        latitude = row['latitude']
        longitude = row['longitude']
        url = f"https://www.immobiliare.it/en/annunci/{row['id']}/?imm_source=homepage"
        mapsgoogle = f"https://www.google.com/maps?q={row['latitude']},{row['longitude']}"
        popup_text = f'<a href="{url}" target="_blank">ID: {row["id"]}</a><br>Price: {row["price"]}<br><a href="{mapsgoogle}" target="_blank">ID: {row["id"]}'
        popup = folium.Popup(popup_text, max_width=250)
        marker = folium.Marker(location=[latitude, longitude], popup=popup, icon=folium.Icon(color=get_color(row['price_category'])))
        marker.add_to(m)

    # Display the map
    return m


min_price_slider = widgets.DateSlider(min=df['observed'].min(), max=df['observed'].max(), value=df['observed'].min(), description='Min Price')
max_price_slider = widgets.DateSlider(min=df['observed'].min(), max=df['observed'].max(), value=df['observed'].max(), description='Max Price')

widgets.interact(update_map, min_price=min_price_slider, max_price=max_price_slider)
```

```python
import folium
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Assuming you have already loaded the DataFrame 'df'
def update_map(min_price, max_price):
    filtered_df = df[(df['price'] >= min_price) & (df['price'] <= max_price)]
    filtered_df['latitude'] = pd.to_numeric(filtered_df['latitude'], errors='coerce')
    filtered_df['longitude'] = pd.to_numeric(filtered_df['longitude'], errors='coerce')
    # Create 5 price categories
    filtered_df['price_category'] = pd.qcut(filtered_df['price'], 5, labels=['very_low', 'low', 'medium', 'high', 'very_high'])

    def get_color(price_category):
        color_map = {
            'very_low': 'green',
            'low': 'lightblue',
            'medium': 'blue',
            'high': 'darkblue',
            'very_high': 'red'
        }
        return color_map[price_category]

    # Create a map centered around the average latitude and longitude of the data
    avg_latitude = filtered_df['latitude'].mean()
    avg_longitude = filtered_df['longitude'].mean()
    m = folium.Map(location=[avg_latitude, avg_longitude], zoom_start=10, width='80%', height=600)

    # Add markers for each property
    for index, row in filtered_df.iterrows():
        latitude = row['latitude']
        longitude = row['longitude']
        url = f"https://www.immobiliare.it/en/annunci/{row['id']}/?imm_source=homepage"
        mapsgoogle = f"https://www.google.com/maps?q={row['latitude']},{row['longitude']}"
        popup_text = f'<a href="{url}" target="_blank">ID: {row["id"]}</a><br>Price: {row["price"]}<br><a href="{mapsgoogle}" target="_blank">ID: {row["id"]}'
        popup = folium.Popup(popup_text, max_width=250)
        marker = folium.Marker(location=[latitude, longitude], popup=popup, icon=folium.Icon(color=get_color(row['price_category'])))
        marker.add_to(m)

    # Display the map
    return m

min_price_slider = widgets.IntSlider(min=df['price'].min(), max=df['price'].max(), value=df['price'].min(), description='Min Price')
max_price_slider = widgets.IntSlider(min=df['price'].min(), max=df['price'].max(), value=df['price'].max(), description='Max Price')

widgets.interact(update_map, min_price=min_price_slider, max_price=max_price_slider)
```
